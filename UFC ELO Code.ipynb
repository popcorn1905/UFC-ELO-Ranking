```python
import pandas as pd

# Load the UFC fighters CSV file
file_path = "C:\\Users\\devom\\Downloads\\ufc-master.csv"
ufc_data = pd.read_csv(file_path)

# Display the first few rows and the columns to understand the structure of the data
ufc_data.head(), ufc_data.columns
```

```python
elo_ratings={}

def expected_result(rating_a, rating_b):
    return 1 / (1 + 10 ** ((rating_b - rating_a) / 400))

def update_elo(winner_rating, loser_rating, result, k=100):
    expected_win = expected_result(winner_rating, loser_rating)
    expected_loss = 1 - expected_win

    winner_new_rating = winner_rating + k * (result - expected_win)
    loser_new_rating = loser_rating + k * ((1 - result) - expected_loss)
    
    return winner_new_rating, loser_new_rating

for index, row in ufc_data.iterrows():
    red_fighter = row['RedFighter']
    blue_fighter = row['BlueFighter']
    winner = row['Winner']

    if red_fighter not in elo_ratings:
        elo_ratings[red_fighter] = 1000
    if blue_fighter not in elo_ratings:
        elo_ratings[blue_fighter] = 1000
        
    red_rating = elo_ratings[red_fighter]
    blue_rating = elo_ratings[blue_fighter]
    
    
    if winner == 'Red':
        red_new_rating, blue_new_rating = update_elo(red_rating, blue_rating, result=1)
    else:
        blue_new_rating, red_new_rating = update_elo(blue_rating, red_rating, result=1)

    elo_ratings[red_fighter] = red_new_rating
    elo_ratings[blue_fighter] = blue_new_rating

elo_ratings_list = [(fighter, rating) for fighter, rating in elo_ratings.items()]
elo_ratings_list
```

```python
fighter_name = "Jared Cannonier"
if fighter_name in elo_ratings:
    print(f"{fighter_name} has an Elo rating of {elo_ratings[fighter_name]:.2f}")
else:
    print(f"{fighter_name} not found in the ratings list.")
```

```python
elo_ratings_list_sorted = sorted(elo_ratings_list, key=lambda x: x[1])

print(f"{'Fighter Name':<30} {'Elo Rating':<10}")
print("="*40)

for fighter, rating in elo_ratings_list_sorted:
    print(f"{fighter:<30} {rating:<10.2f}")
```

```python
fight_count = {}

for index, row in ufc_data.iterrows():
    red_fighter = row['RedFighter']
    blue_fighter = row['BlueFighter']

    if red_fighter not in fight_count:
        fight_count[red_fighter] = 0
    if blue_fighter not in fight_count:
        fight_count[blue_fighter] = 0

    fight_count[red_fighter] += 1
    fight_count[blue_fighter] += 1

elo_ratings_filtered = [ (fighter, rating) for fighter, rating in elo_ratings_list if fight_count.get(fighter, 0) >= 5 ]

elo_ratings_filtered_sorted = sorted(elo_ratings_filtered, key=lambda x: x[1])

print(f"{'Fighter Name':<30} {'Elo Rating':<10}")
print("="*40)
for fighter, rating in elo_ratings_filtered_sorted:
    print(f"{fighter:<30} {rating:<10.2f}")
```

```python
elo_ratings_list_sorted_desc = sorted(elo_ratings_list, key=lambda x: x[1], reverse=True)

print(f"{'Fighter Name':<30} {'Elo Rating':<10}")
print("="*40)

for fighter, rating in elo_ratings_list_sorted_desc:
    print(f"{fighter:<30} {rating:<10.2f}")
    
```
